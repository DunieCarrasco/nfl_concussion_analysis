```python
city = "Glendale, Arizona"

# Build the endpoint URL
url = f"https://api.geoapify.com/v1/geocode/search?text={city}&format=json&apiKey={geoapify_key}"
```

```python
# Run a request to endpoint and convert result to json
geo_data2 = requests.get(url).json()

# Print the json
print(json.dumps(geo_data2, indent=4, sort_keys=True))
```

```python

lat = geo_data2["results"][0]["lat"]
lon = geo_data2["results"][0]["lon"]
lat 
lon
```

```python
coordinates2 = {
    "latitude": [lat],
    "longitude": [lon],
    "City": ["Glendale, Arizona"],
    "concussion": [1]
}
coordinates_df = pd.DataFrame(coordinates2)
```
